```python
#importing lib plus setup reading for csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#setting up the dataets
'''
    df1 contains 
    movieID (unique for each movie)
    cast (each actors/actress including supporting artists)
    crew (including te directors, writer, editors, etc)
'''
df1 = pd.read_csv('./tmdb-movie-metadata/tmdb_5000_credits.csv')

'''
    df2 contains 
    budget, genre, hompage, 
    id, keywords, original_language,
    original_title, overview, popularity
    prouctioncompanies, production_counrties,
    release_date, revenue, runtime, status, tagline
    title, vote_avg, vote_count
'''
df2 = pd.read_csv('./tmdb-movie-metadata/tmdb_5000_movies.csv')
```

```python
#mergig the two datsets based off of id
df1.columns = ['id', 'title', 'cast', 'crew']
df2 = df2.merge(df1, on='id')
```

```python
#test to see merge
df2.head(5)
```

```python
#calulating a weigted rating
#C is the mean vote across the whole report
C = df2['vote_average'].mean()
C
```

```python
#m is the minimum votes needed to be listed in the chart
m= df2['vote_count'].quantile(0.9)
m
```

```python
#filtering moivies based off of which qualify from rating
q_movies = df2.copy().loc[df2['vote_count'] >= m]
q_movies.shape
```

```python
#weighted rating function
def weighted_rating(X, m=m, C=C):
    V = X['vote_count'] #number of votes for the movie
    R = X['vote_average'] #average rating of the movie
    
    #calculations based off of WR formula
    return ((V/(V+m))*R) + ((m/(m+V))*C)
```

```python
#defining a new score and calculating its value from WR
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
#sorting the movies based off of score
q_movies = q_movies.sort_values('score', ascending=False)

#printig top movies
q_movies[['original_title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
#sort by popularity 
pop = df2.sort_values('popularity', ascending=False)

#drawing bar graph
plt.figure(figsize=(12,4))
plt.barh(pop['original_title'].head(6), pop['popularity'].head(6), align='center', color='lightcoral')
plt.gca().invert_yaxis()
plt.xlabel("Popularity")
plt.title("Popular movies")
```
